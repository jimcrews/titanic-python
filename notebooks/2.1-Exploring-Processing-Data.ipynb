```python
#imports
import pandas as pd
import numpy as np
import os
```

# Import Data

```python
# set the path of the raw data
raw_data_path = os.path.join(os.path.pardir, 'data', 'raw')
train_file_path = os.path.join(raw_data_path, 'train.csv')
test_file_path = os.path.join(raw_data_path, 'test.csv')

# read the data with all default parameters
train_df = pd.read_csv(train_file_path, index_col='PassengerId')
test_df = pd.read_csv(test_file_path, index_col='PassengerId')

# Add Survived to 'test' so as to match columns with 'train' and concat
test_df['Survived'] = -888

df = pd.concat((train_df, test_df),axis=0)
```

```python
df
```

# Distributions

### Age

```python
# use hist to create a histogram, using the default amount of bins
df.Age.plot(kind='hist', title='histogram for Age', color='c', rwidth=0.9)
```

```python
# use bins to add or remove bins
df.Age.plot(kind='hist', title='histogram for Age', color='c', rwidth=0.9, bins=30)
```

```python
# kde plot requires scipy. install from anaconda cmd
# conda install scipy

# use kde for a density plot
df.Age.plot(kind='kde', title='density plot for Age', color='c');
```

### Fare

```python
df.Fare.plot(kind='hist', title='histogram for Fare', color='c', bins=20)
```

```python
# compare skewness between age and fare

print('skewness for age: {0:.2f}'.format(df.Age.skew()))
print('skewness for fare: {0:.2f}'.format(df.Fare.skew()))
```

# Scatter Plots for Bivariate distribution

```python
# .. no correlation
df.plot.scatter(x='Age', y='Fare', color='c', title='scatter plot : Age vs Fare')
```

```python
# use alpha to set the transparency.. no correlation
df.plot.scatter(x='Age', y='Fare', color='c', title='scatter plot : Age vs Fare', alpha=0.1)
```

```python
# check correlation between class and fare
df.plot.scatter(x='Pclass', y='Fare', color='c', title='scatter plot : Age vs Fare', alpha=0.15)
```
